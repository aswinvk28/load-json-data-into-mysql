```python
import os
import sqlalchemy
from tqdm import tqdm
import sys
import re
import json
```

```python
# file name for importing JSON
input_filename = os.environ.get('input_filename', 'Software-Cleaned.json')

# file data storage location
data_dir = os.environ.get('data_dir', './data')

# url for mysql
url = os.environ.get("url", "mysql+pymysql://admin:admin@192.168.0.21:3306/amazon_reviews")
```

```python
from sqlalchemy import create_engine, event

from sqlalchemy.orm import sessionmaker
engine = create_engine(url, future=True)
Session = sessionmaker(bind=engine)

conn = engine.connect()
```

```python
from sqlalchemy import Table, MetaData, Column, Integer, String, TIMESTAMP, SmallInteger, Text, Date

metadata = MetaData()

software_reviews = Table(
    "software_reviews",
    metadata,
    Column('id', Integer, primary_key=True),
    Column('overall', SmallInteger),
    Column('vote', String(10)),
    Column('verified', SmallInteger),
    Column('reviewTime', Date),
    Column('reviewerID', String(255)),
    Column('asin', String(255)),
    Column('style', Text),
    Column('reviewerName', String(255)),
    Column('reviewText', Text),
    Column('summary', Text),
    Column('unixReviewTime', Integer)
)
```

```python
# open the input file
with open(os.path.join(data_dir, input_filename), 'r') as file:
    conn.execution_options(isolation_level="AUTOCOMMIT")
    for ii,line in tqdm(enumerate(file)):
        json_line = json.JSONDecoder().decode(line)
        json_line['style'] = json.JSONEncoder().encode(json_line['style'] if 'style' in json_line else {})
        conn.execute(
            software_reviews.insert(),
            json_line
        )
        if ii % 100 == 0:
            conn.commit()
conn.commit()
print("No of records: ", ii+1)
```
